Bei der NVDA Aktie kann man eine starke Korrelation zwischen dem Kurs und den Google Trends erkennen. Bei den anderen beiden Aktien ist die Korrelation nicht so stark. Dies könnte daran liegen, dass die Google Trends für "Google stock" und "Microsoft stock" nicht so aussagekräftig sind wie für "NVIDIA stock".

Für die Google und Microsoft Aktien sollten andere Suchbegriffe (z.B. "buy Microsoft", "sell Google") benutzt werden um eine bessere Korrelation zu erreichen.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Notebook-spezifische Einstellungen (optional)
%matplotlib inline
sns.set_style("whitegrid")

# 1) Ticker-Liste und Datenstrukturen
tickers = ["NVDA", "GOOG", "MSFT"]
stock_data_dict = {}

# 2) Aktienkurs-CSV einlesen
for ticker in tickers:
    file_path = f"../../03_Daten/processed_data/historical_stock_data_weekly_{ticker}_flat.csv"
    df = pd.read_csv(file_path, parse_dates=["Date"], index_col="Date")
    stock_data_dict[ticker] = df

    print(f"\n--- {ticker} Aktienkurse (head) ---")
    display(df.head())

# 3) Google Trends-Daten einlesen
trends_data = pd.read_csv("../../03_Daten/raw_data/google_trends_weekly.csv",
                          parse_dates=["date"], index_col="date")
print("\n--- Google Trends (head) ---")
display(trends_data.head())

# 4) Explorative Datenanalyse pro Ticker
for ticker in tickers:
    print(f"\n=== EDA für {ticker} ===")
    # Passende Spalte für Google Trends ermitteln
    if ticker == "NVDA":
        trends_col = "NVIDIA stock"
    elif ticker == "GOOG":
        trends_col = "Google stock"
    elif ticker == "MSFT":
        trends_col = "Microsoft stock"

    # Merge nach Datumsindex (innerer Join)
    merged_data = stock_data_dict[ticker].merge(trends_data, left_index=True, right_index=True, how='inner')

    # 4.1 Beschreibende Statistiken & Fehlende Werte
    print("\nBeschreibende Statistiken:")
    display(merged_data.describe())

    print("Fehlende Werte pro Spalte:")
    display(merged_data.isnull().sum())

    # 4.3 Rendite berechnen und plotten (pct_change)
    merged_data[f'{ticker}_Return'] = merged_data['Close'].pct_change()
    plt.figure(figsize=(12, 4))
    plt.plot(merged_data.index, merged_data[f'{ticker}_Return'], label=f'{ticker} wöchentliche Rendite')
    plt.title(f"{ticker}: Wöchentliche Renditen")
    plt.xlabel("Datum")
    plt.ylabel("Rendite")
    plt.legend()
    plt.show()

    # 4.4 Korrelation und Heatmap
    corr_matrix = merged_data.corr()
    print("\nKorrelationsmatrix:")
    display(corr_matrix)

    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f"{ticker}: Korrelationsmatrix")
    plt.show()

    # 4.5 Histogramm Schlusskurs
    plt.figure(figsize=(8, 4))
    sns.histplot(merged_data['Close'].dropna(), kde=True)
    plt.title(f"{ticker}: Verteilung des Schlusskurses")
    plt.xlabel("Close")
    plt.show()

    # 4.6 Boxplot Schlusskurs
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=merged_data['Close'].dropna())
    plt.title(f"{ticker}: Boxplot des Schlusskurses")
    plt.xlabel("Close")
    plt.show()
```

NVDA:
Im Plot von NVDA ist ein deutlicher langfristiger Aufwärtstrend im Schlusskurs zu erkennen, begleitet von signifikanten Peaks im aggregierten Google Trends Suchinteresse (Trend Average). Die geglättete Linie hebt den allgemeinen Anstieg des Suchinteresses hervor und zeigt, dass Phasen mit starkem Kursanstieg oftmals mit einem erhöhten Anlegerinteresse einhergehen. Diese enge Synchronizität zwischen Kursbewegungen und Suchvolumen deutet darauf hin, dass Nachrichten und unternehmensspezifische Ereignisse einen starken Einfluss auf die Marktstimmung bei NVIDIA haben.

GOOG:
Der GOOG-Plot zeigt einen moderateren Verlauf sowohl in den Schlusskursen als auch im aggregierten Suchinteresse. Die Schwankungen im Google Trends Suchvolumen sind zwar erkennbar – insbesondere während spezieller Ereignisse oder Publikationen – jedoch ist die Korrelation mit den Kursbewegungen insgesamt weniger ausgeprägt als bei NVDA. Dies legt nahe, dass das allgemeine Suchverhalten für Google zwar bestimmte Marktphasen signalisiert, aber nicht so stark direkt den Kurs beeinflusst, was auf eine komplexere Wechselwirkung mit anderen Einflussfaktoren hindeutet.

MSFT:
Bei MSFT sind sowohl die Schlusskurse als auch die aggregierten Google Trends-Daten relativ moderat und weisen moderate Schwankungen auf. Die aggregierte Trend-Linie, sowohl als einfacher Durchschnitt als auch als geglättete Darstellung, folgt im Großen und Ganzen den Kursbewegungen, zeigt aber weniger ausgeprägte Spitzen als bei NVDA. Dies deutet darauf hin, dass das Suchinteresse bei Microsoft zwar einen gewissen Zusammenhang mit den Kursen aufweist, jedoch durch weitere Faktoren vermittelt oder abgeschwächt wird, sodass der direkte Einfluss des Anlegerinteresses weniger intensiv ist als bei NVIDIA.

**Beobachtungen**
- **NVDA:** Die aggregierte Trend‑Average‑Linie (orange) und ihr geglättetes Pendant (grün) steigen ab 2023 deutlich an und korrelieren gut mit dem starken Kursanstieg von NVIDIA. Vor allem im steilen Anstiegs­segment folgt das Suchinteresse dem Kursverlauf, wenn auch leicht gedämpft.

- **MSFT:** Microsoft zeigt moderate, aber verlässliche Anstiege im Suchinteresse zu bestimmten Zeitpunkten (z. B. Produkt‑Launches), die tendenziell mit Kursbewegungen harmonieren. Die Korrelation ist stärker als bei Google, bleibt aber hinter NVIDIA zurück.

**Der Code musste angepasst werden:**

Ursprünglich lagen die Google‑Trends‑Daten **monatlich**, während die Aktien‑ und RSI‑Daten **wöchentlich** (z.B. Donnerstags) vorliegen. In der ersten Plot‑Version sprachen wir also aneinander vorbeilaufende Zeitreihen an – die X‑Achsen waren schlicht unterschiedlich getaktet.
Um valide Vergleiche und Korrelationen zu ermöglichen, habe ich:
1. den exakten Wochentag der Aktien‑Zeitreihe automatisch ermittelt (z. B. “THU”),
2. die monatlichen Trends per `.resample("W-THU").interpolate()` in echte Wochenwerte überführt,
3. und erst danach Aggregation und Merge durchgeführt.

So stimmen Frequenzen, Indizes und Zeitpunkte jetzt exakt überein, und die Trend‑Linien können fair mit den wöchentlichen Kursen verglichen werden.
